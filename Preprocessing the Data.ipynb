```python
import pandas as pd
raw_csv_data = pd.read_csv('Absenteeism-data.csv')
df_copy = raw_csv_data.copy()
df_copy.drop('ID', axis = 1, inplace = True)
reasons_column = pd.get_dummies(df_copy['Reason for Absence'])
reasons_column['check'] = reasons_column.sum(axis = 1)
reasons_column.head()
reasons_column['check'].sum(axis = 0)
reasons_column['check'].unique()
reasons_column = reasons_column.drop(['check'], axis = 1)
reasons_column.head()
```

## There are 3 steps to take before finalizing our work on this column:

##    Step 1: With the current state of the DF, if we decide to add all the dummy variables to it, we would observe multicollinearity. So drop the "Reason for Absence"

```python
df_copy = df_copy.drop('Reason for Absence', axis = 1)
```

```python
df_copy.head()
```

## Step 2: Grouping the variables:
    - If we add the variables in the DF, there would be almost 40 columns.. (Too many to look at clearly)
    - So let's group to clean up the DF:

### Creating a new DF object for each grouping:
    - Class/Grouping 1: Columns 1-14 for diseases
    - Class/Grouping 2: Columns 15-17 for pregnancy 
    - Class/Grouping 3: Columns 18-21 
    - Class/Grouping 4: Columns 22-28 
    

- The 'Check' column is to verify that there was only one reason for the absence of a person
- This allows us to proceed by:
    - Replacing entire row of values 1-14 (Reason_1) and so on with: 
        - 0 if: none of the values on the given row were eual to 1
            - This means that the reason of absence was not in this group
        - 1 if: somewhere among these 14 columns we have observed the number 1
              - This means that the reason of absence was in this group
    

```python
reason_columns['check'] = reason_columns.sum(axis = 1)
```

```python
reason_columns.check
```

```python
# Below will return a series:
reason_columns.iloc[:, 0:14].max(axis=1)
```

```python
reason_type_1 = reason_columns.iloc[:, 0:14].max(axis=1)
reason_type_2 = reason_columns.iloc[:, 14:17].max(axis=1)
reason_type_3 = reason_columns.iloc[:, 17:21].max(axis=1)
reason_type_4 = reason_columns.iloc[:, 21:].max(axis=1)
```

```python
reason_type_1.head()
```

```python
reason_type_2.head()
```

## STEP 3:
####    - In previous section: steps 1 and 2 need to be done during preprocessing for Reason for Absence column
####    - Now, add newly created and simply organzied 'reason_columns' into main DF:

```python
df_copy = pd.concat([df_copy, reason_type_1, reason_type_2, 
                     reason_type_3, reason_type_4], axis = 1)
```

## Working with Days of the Week:
### - Create the day of the week column (with numbers 0-6, built in values)
### - Create weekday_generator method:

```python
def weekday_maker(date_value):
    return date_value.weekday()
```

```python
weekday_obj = df_reason_mod['Date'].apply(weekday_maker)
```

```python
df_reason_mod['Day of the Week'] = weekday_obj
```
